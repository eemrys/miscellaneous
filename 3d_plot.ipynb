```python
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def z_function(x, y):
    return np.exp(-np.sqrt(x**2+y**2)/4)*np.sin(3*np.sqrt(x**2+y**2))
```

```python
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
```

```python
X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)
```

```python
graph = plt.figure()
axes = plt.axes(projection="3d")
axes.plot_wireframe(X, Y, Z, color='green')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_zlabel('z')
axes.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='winter')
axes.set_title('surface');
plt.show()
```
